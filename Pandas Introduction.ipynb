## Pandas Objects
        Pandas objects are advanced versions of NumPy structured arrays in which the rows and columns are identified with    labels instead of simple integer indices

### 1) Series: 
        		-one dimensional labelled array.
                -supports different datatypes like integer, float, string etc

                Syntax: pd.Series(data, index, dtype)
                      data – It can be a list, list of lists or even a dictionary.
                      index – can be explicitly defined for different valuesif required.
                      dtype – represents the data type used in the series (optional parameter).

```python
int_series = pd.Series(data = [78, 92, 36, 64, 89])  
print(int_series)             #-5  -4  -3  -2  -1
```

```python
str_series = pd.Series(data = ['Jack', 'John', 'Anthony', 'Akash', 'Ron'], index=[4,3,2,1,0] )  
str_series
```

```python
mixed_series = pd.Series(data = [75, 10, 'Math', 'Science', 55.12], index=['a', 'b', 'c', 'd', 'e'])
mixed_series
```

### 3) Masking
        -replaces values where the condition is True
        
        Syntax: DataFrame.mask(cond, other = nan, inplace =, axis = 0)

```python
marks = [{'Chemistry': 67, 'Physics': 45, 'Mathematics': 50, 'English' : 19},
         {'Chemistry': 90, 'Physics': 92, 'Mathematics': 87, 'English' : 90}, 
         {'Chemistry': 66, 'Physics': 72, 'Mathematics': 81, 'English' : 72}, 
         {'Chemistry': 32, 'Physics': 40, 'Mathematics': 12, 'English' : 68}]
marks_df = pd.DataFrame(marks, index = ['Subodh', 'Ram', 'Abdul', 'John'])
marks_df
```

```python
marks_df < 33
```

```python
# suppose we do not want to reveal the marks of students who have failed.
f = marks_df < 33
marks_df.mask(f, 'Fail',inplace=True)
```

```python
marks_df
```

### 5) Preserving index
            -Pandas preserves the index and column labels in the output
            -automatically align indices when passing the objects to the functions

```python
marks = [{'Chemistry': 67, 'Physics': 45, 'Mathematics': 50, 'English' : 19},
         {'Chemistry': 90, 'Physics': 92, 'Mathematics': 87, 'English' : 90}, 
         {'Chemistry': 66, 'Physics': 72, 'Mathematics': 81, 'English' : 72}, 
         {'Chemistry': 32, 'Physics': 40, 'Mathematics': 12, 'English' : 68}]
marks_df = pd.DataFrame(marks, index = ['Subodh', 'Ram', 'Abdul', 'John'])
marks_df
```

```python
#encrypting marks as sine of marks
encrypted_marks = np.sin(marks_df)
encrypted_marks
```

```python
#In case of a requirement where the index has to be restored to the default index
encrypted_marks.reset_index(inplace = True,drop=True)
encrypted_marks
```

### 6) Broadcasting operation
        -refers to a set of rules to operate between data of different sizes and shapes

```python
marks_df
```

```python
#Supppose we want to award five bonus marks to all the students
new_marks = marks_df + 5
new_marks
```

```python
#Adding different marks to different subjects
new_marks = marks_df + [5,10,10,2]
new_marks
```

```python
# Apply Function :  used to apply a function along an axis of the DataFrame
# Syntax: DataFrame.apply(func, axis = 0, result_type = None)

#to get the total marks scored in "each subject".
marks_df.apply(np.sum, axis = 0)      
```

```python
#to get the total marks scored by "each student".
marks_df.apply(np.sum, axis = 1)  
```

```python
#Suppose students were unable to attend the next exam of Biology due to the pandemic. 
#Hence, we decide to award them average marks based on their previous performance.
marks_df['Biology'] = marks_df.apply(np.mean, axis=1)
marks_df
```

### 7) Aggregation
            - process of turning the values of a dataset into one single value
             (or) simply getting a single value output by operating on whole dataset
             
             Syntax : DataFrame.agg(func, axis = 0)

```python
#aggregating minimum and maximum marks for each student
marks_df.agg(['min', 'max'], axis=1)
```

```python
#calculating total marks for each student
marks_df.agg(['sum','mean'], axis=1)
```

### 11) Pivot Table
            - used to summarise, sort, reorganise, group, count, total or average data stored in a table.
            Syntax: pd.pivot_table(data, index, aggfunc)

## Visualizing Data using Panda Plots
            -Syntax:  df.plot(X, y, marker, kind)
                       x = value on X axis
                       y = value on y axis
                       marker = shape in case of specific plots like a scatter plot
                       kind = type of plot

```python
#A scatter plot to visualize the relationship between weight and mpg.
df.plot(x = 'weight', y = 'mpg', kind = 'scatter',marker='*')
```

```python
#A bar plot to visualize mean acceleration in different years.
df.groupby('model_year').mean()[['acceleration']].plot(kind = 'bar')
```

```python
#A bar plot to visualize the sorted mean values of acceleration with respect to number of cylinders.
df.groupby('cylinders').mean().sort_values('acceleration')[['acceleration']].plot(kind = 'bar')
```
