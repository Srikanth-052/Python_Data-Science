## Introduction
          - most basic Python library used for visualizing data.
          - Also built on top of Numpy
          - matplotlib.pyplot is used for generating 2D graphics in python
          
          Syntax:
          plt.plot(x, y, color=’green’, marker=’o’, linestyle=’dashed’, linewidth=2, markersize=12)

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

### Creating a simple plot

```python
#creating two arrays
X = np.arange(0,10,0.1)
X
Y = np.sin(X)
```

```python
#plotting the values
plt.plot(X,Y)
```

            - matplotlib.pyplot creates the default elements like Figure and Axes required and then plots the data for us.
            - This method is not giving much control over the plots and its features. 

##### Figure -> Canvas
##### Axes -> Bounding Box
##### Plot -> actual plot
##### Lables, title, legend, annotations etc -> Canvas

### Object oriented approach for Ploting
 		-gives more control over customization of the plots.

		-Steps Involved-
         Creating a figure object: 
		 Setting up the axes: [left, bottom, width, height]
		 Creating a plot using the axes object: 
		 Creating multiple plots using the same axes object
		 Setting up the title, label, and legend for a plot             

```python
#creating sample data for plotting
x = np.arange(1,16)   
y = x**2   
z = (y/2)+40    
```

```python
# STEP 1 :- creating an empty canvas/figure
fig = plt.figure()

#STEP 2 :- setting axes as [left, bottom, width, height]
ax = fig.add_axes([0,0,1,1])         #notice we are using "fig" object which is storing our canvas

# #STEP 3 :- Creating a plot using the axes object
ax.plot(x, y, color="g",label='line1' )
ax.plot(x, z, color="r", label='line2')  #simply plotting another line on the same axes

# #STEP 4 :- Setting up the title, label, and legend for a plot
ax.set_title('Title of the Plot',fontsize=15)    #setting title
ax.set_xlabel('Label for X axis',fontsize=15)    # setting X label
ax.set_ylabel('Label for Y axis',fontsize=15)    # setting Y _label
ax.legend()    #adding legend
```

### Creating multiple axes

```python
#creating an empty canvas/figure
new_fig = plt.figure(figsize=(8,5))

#setting axes as [left, bottom, width, height]
ax1 = new_fig.add_axes([0,0,1,1])
ax2 = new_fig.add_axes([0.10,0.40,0.45,0.45])

#plotting the lines
ax1.plot(x,y,color="g", label='line1' )
ax1.set_title('Title for Plot 1',fontsize=15)    #setting title
ax1.set_xlabel('Plot 1 - X axis',fontsize=15)    # setting X label
ax1.set_ylabel('Plot 1 - Y axis',fontsize=15)    # setting Y _label
ax1.legend()    #adding legend

ax2.plot(x,z,color="r", label='line2')
ax2.set_title('Title for Plot 2',fontsize=15)    #setting title
ax2.set_xlabel('Plot 2 - X axis',fontsize=15)    # setting X label
ax2.set_ylabel('Plot 2 - Y axis',fontsize=15)    # setting Y _label
ax2.legend()    #adding legend
```

### Subplots
        - used to create a common layout for multiple plots
        - 'm(rows) x n(columns)' number of plots can be created using a subplot
        
        Syntax:  plt.subplots(rows, cols)

```python
#creating an empty canvas/figure
#subplot_fig = plt.figure(figsize=(10,30))

#initializing two axes for the subplot
subplot_fig, (ax1, ax2) = plt.subplots(1, 2)  #intializing Subplots  

ax1.plot(x,y,color="g", label='line1' )
ax1.set_title('Title for Plot 1',fontsize=15)    #setting title
ax1.set_xlabel('Plot 1 - X axis',fontsize=15)    # setting X label
ax1.set_ylabel('Plot 1 - Y axis',fontsize=15)    # setting Y _label
ax1.legend()

ax2.plot(x,z,color="r", label='line2')
ax2.set_title('Title for Plot 2',fontsize=15)    #setting title
ax2.set_xlabel('Plot 2 - X axis',fontsize=15)    # setting X label
ax2.set_ylabel('Plot 2 - Y axis',fontsize=15)    # setting Y _label
ax2.legend()

# ax3.plot(y,z,color='b', label='line3' )
# ax3.set_title('Title for Plot 3',fontsize=15)    #setting title
# ax3.set_xlabel('Plot 3 - X axis',fontsize=15)    # setting X label
# ax3.set_ylabel('Plot 3 - Y axis',fontsize=15)    # setting Y _label
# ax3.legend()

subplot_fig.tight_layout()
```

### 1) Box Plot
            -good indication of distribution of data about the median
            -standardized way of displaying the distribution of data based on the five-number summary

```python
#creating an empty canvas/figure
box_fig = plt.figure(figsize=[8,5])

#setting axes
ax = box_fig.add_axes([0, 0, 1, 1])

#plotting boxplot
ax.boxplot(df['mpg'])

#aetting Title
ax.set_title('Distribution of mileage')
```

```python
df.describe()
```

```python
#processing data for city mileage
df['city_mileage']=df['mpg']*0.75
df.head()
```

```python
#Comparing avarage mpg and city mpg using boxplot
mpg_list = [df['mpg'],df['city_mileage']]

#creating an empty canvas/figure
box_fig = plt.figure(figsize=[8,5])

#setting axes
ax = box_fig.add_axes([0, 0, 1, 1])

#plotting boxplot
ax.boxplot(mpg_list, widths=0.5)

#setting Title
ax.set_title('Distribution of Average MPG vs City MPG')
ax.set_ylabel('Mileage per gallon')
```

### 2) Scatter plot
            -uses dots or markers to represent a value on the axes
            -one of the simplest plot which can accept both quantitative and qualitative values
            -identifying the type of correlation between variables

```python
#Suppose we want to infer the correlation between horsepower and mileage.

#creating an empty canvas/figure
scatter_fig = plt.figure(figsize=[10,5])

#setting axes
ax = scatter_fig.add_axes([0, 0, 1, 1])

#plotting scatter plot
ax.scatter(df['horsepower'],df['mpg'], marker='*', color='r')

#Setting lables and Titles
ax.set_title("Scatter plot of horsepower and mileage",fontsize=16)
ax.set_ylabel('Miles per gallon',fontsize=12)
ax.set_xlabel('Horsepower',fontsize=12)
```

```python
#finding correlation between mileage and horsepower based on the origin
#Which country’s car has the highest mileage?
#Which country’s car has the highest horsepower?

#creating an empty canvas/figure
scatter_fig = plt.figure(figsize=[12,5])

#setting axes
ax = scatter_fig.add_axes([0, 0, 1, 1])

#assigning different colors for each origin
origin_list = df["origin"].unique()
#origin_list
colors = ["r","g","b"]

# plotting based on the origin
for origin,color in zip(origin_list,colors):
    x = df[df["origin"] == origin]["horsepower"]
    y = df[df["origin"] == origin]["mpg"]
    plt.scatter(x,y,color = color,label=origin)

ax.set_title("Scatter plot of Horsepower and Mileage based on Origin",fontsize=16)
ax.set_ylabel('Mileage per gallon',fontsize=12)
ax.set_xlabel('Horsepower',fontsize=12)
plt.legend()

```

### 3)  Bar Chart:
            -graph with rectangular bars that usually compare different categories
            -length of bar indicates total number of values in that category


```python
#We want to know the number of cars released in each year

#Creating a grouped DataFrame according to model year
grouped_df = df.groupby(['model_year']).count()[['name']]

grouped_df.reset_index(inplace = True)
grouped_df.head()
```

```python
#creating an empty canvas/figure
box_fig = plt.figure(figsize=[7,5])

#setting axes
ax = box_fig.add_axes([0, 0, 1, 1])

#plotting bar graph
ax.bar(grouped_df['model_year'], grouped_df['name'])

#Setting titles and lables
ax.set_title("Number of cars in each year",fontsize=16)
ax.set_ylabel('Number of cars',fontsize=12)
ax.set_xlabel('Model year',fontsize=12)
```

### 4) Histogram
            -used for continuous data
            -height of the bar is based on the number of values that fall into the corresponding intervals
            -ideally suited to obtain the frequency distribution of a given data

```python
# want to identify the frequency distribution of horsepower.

#creating an empty canvas/figure
hist_fig = plt.figure(figsize=[8,6])

#setting axes
ax = hist_fig.add_axes([0, 0, 1, 1])

#plotting histogram
ax.hist(df['horsepower'], bins=20) #bins = defines the number of equal-width bins in the range

#setting lables and title
ax.set_title('Distribution of Horsepower',fontsize = 17)
ax.set_ylabel('Frequency',fontsize=12)
ax.set_xlabel('Horsepower',fontsize=12)
```

### 5) Pie Chart
            -divides the entire dataset into distinct groups
            - circle split into pies and each pie represents a group	
            -size of the pie is proportional to the number of items in each group compared to others

```python
#Suppose we want the details of origin of the cars is distributed  

pie_df = pd.DataFrame()    #creating a sub dataframe to plot the pie chart
pie_df['Count'] = df['origin'].value_counts()    #getting the count of ‘origin’ and assigning to the df
pie_df
```

```python
pie_df = pie_df.reset_index()    #re-arranging the index
pie_df.rename(columns={'index':'Country'},inplace=True)    #re-naming the col
pie_df
```

```python
#creating an empty canvas/figure
pie_fig = plt.figure(figsize=[8,6])

#setting axes
ax = pie_fig.add_axes([0, 0, 1, 1])

#plotting pie chart
ax.pie(pie_df['Count'], labels=pie_df['Country'])

ax.set_title('Origin of the Cars')
```

```python
fig = plt.figure(figsize=[8,6])
ax = fig.add_axes([0, 0, 1, 1])

#explode-To get an elevated view for the selected pie. 
explode = [0.1, 0.1, 0.1]


#colors-To customize the colors for the plot.
colors = ['yellowgreen', 'mediumaquamarine', 'khaki']

ax.set_title('Origin of the Cars')

#autopct-To add the percentage of the distribution in the pie chart.
#shadow-To add shadow to the plot.
#startangle- To change the starting angle of the pie chart.

ax.pie(pie_df['Count'], labels=pie_df['Country'],
        autopct='%0.1f%%',explode = explode, 
        shadow = True, startangle = 150,
        colors = colors)
```
